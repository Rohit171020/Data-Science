```python
# plot y = sin(x) and y = cos(x) for 0 to 2*pi. use dashed like for sin(x)
# and solid lines for cos(X)
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 2*np.pi, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = 'y = sin(x)', linestyle = '--', color = 'blue')
plt.plot(x, y2, label = 'y = cos(x)', linestyle = '-', color = 'red')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Trigonometric functions')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values for both functions
x = np.linspace(0.1, 5, 100)  # Start at 0.1 to avoid issues with ln(x)

# Compute y values
y1 = np.log(x)  # Natural logarithm
y2 = np.exp(x)  # Exponential

# Plot both functions
plt.plot(x, y1, label='y = ln(x)', color='red', linestyle='-.')
plt.plot(x, y2, label='y = e^x', color='green', linestyle='-')

# Add labels, legend, and grid
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Exponential vs Logarithmic Functions (Shared x)')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
